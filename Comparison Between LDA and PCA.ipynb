```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import pandas as pd
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
dataset = datasets.load_iris()
```

```python
target_names       = np.array(dataset.target_names)
target             = np.array(dataset.target)
data               = np.array(dataset.data)
feature_names      = np.array(dataset.feature_names)
```

```python
pca = PCA(n_components=2)
lda = LinearDiscriminantAnalysis(n_components=2)
```

```python
data_reduced_pca = pca.fit(data).transform(data)
data_reduced_lda = lda.fit(data, target).transform(data)
```

```python
data_reduced_lda.shape
```

```python
colors = ['blue','green','red']
fig, axs = plt.subplots(1,2,figsize=(20,5))
for target_name, i,color in zip(target_names, [0,1,2], colors):
    axs[0].scatter(data_reduced_pca[target==i,0], data_reduced_pca[target==i,1], c=color, linewidths=2, label=target_name)
    axs[0].set_title('PCA - Principle Component Analysis')
    axs[0].set_xlabel('PC1')
    axs[0].set_ylabel('PC2')
    axs[0].legend()
for target_name, i,color in zip(target_names, [0,1,2], colors):
    axs[1].scatter(data_reduced_lda[target==i,0], data_reduced_lda[target==i,1], c=color, linewidths=2, label=target_name)
    axs[1].set_title('LDA - Linear Discriminant Analysis')
    axs[0].set_xlabel('PC1')
    axs[0].set_ylabel('PC2')
    axs[1].legend()
```
